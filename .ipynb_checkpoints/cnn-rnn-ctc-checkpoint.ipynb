```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from tqdm import tqdm

# Define transforms
image_transforms = transforms.Compose([
    transforms.Resize((32, 128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

class UrduOCRDataset(Dataset):
    def __init__(self, df, char_to_idx, transform=None):
        self.df = df.reset_index(drop=True)
        self.char_to_idx = char_to_idx
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # Directly use the image path without adding 'test/'
        img_path = self.df.loc[idx, 'image_path']  
        image = Image.open(img_path).convert('RGB')  # Open image directly from the path

        if self.transform:
            image = self.transform(image)

        label = self.df.loc[idx, 'label']
        label_encoded = [self.char_to_idx[c] for c in label]
        label_tensor = torch.tensor(label_encoded, dtype=torch.long)

        return image, label_tensor, len(label_encoded)  # Return label length too
        
# Custom collate_fn for DataLoader
def ocr_collate_fn(batch):
    images, labels, label_lengths = zip(*batch)
    images = torch.stack(images)
    labels = torch.cat(labels)
    label_lengths = torch.tensor(label_lengths, dtype=torch.long)
    return images, labels, label_lengths

# Example: Create datasets and dataloaders
# train_df, val_df, char_to_idx should already be defined
train_dataset = UrduOCRDataset(train_df, char_to_idx, transform=image_transforms)
val_dataset = UrduOCRDataset(val_df, char_to_idx, transform=image_transforms)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=0, collate_fn=ocr_collate_fn)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=0, collate_fn=ocr_collate_fn)

# Visual check
batch = next(iter(train_loader))
images_batch, labels_batch, label_lengths_batch = batch
plt.imshow(images_batch[0].permute(1, 2, 0))
plt.title("Sample Image from Batch")
plt.axis("off")
plt.show()

# Optional: decode the first label
decoded_label = ''.join([idx_to_char[idx.item()] for idx in labels_batch[:label_lengths_batch[0]]])
print(f"✅ Decoded Urdu Label: {decoded_label}")

# Define the model
class UrduOCRModel(nn.Module):
    def __init__(self, num_classes):
        super(UrduOCRModel, self).__init__()
        self.cnn = nn.Sequential( 
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(128, 256, 3, padding=1), nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1), nn.ReLU(), nn.MaxPool2d((2, 1)),
            nn.Conv2d(256, 512, 3, padding=1), nn.BatchNorm2d(512), nn.ReLU(), nn.MaxPool2d((2, 1)),
            nn.Conv2d(512, 512, 2), nn.ReLU(),
        )
        self.rnn = nn.LSTM(512, 256, num_layers=2, bidirectional=True, batch_first=True)
        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        conv_out = self.cnn(x)            # [B, 512, 1, W]
        conv_out = conv_out.squeeze(2)    # [B, 512, W]
        conv_out = conv_out.permute(0, 2, 1)  # [B, W, 512]
        rnn_out, _ = self.rnn(conv_out)
        output = self.fc(rnn_out)         # [B, W, num_classes]
        return output.permute(1, 0, 2)    # [W, B, num_classes] for CTC Loss

# Instantiate model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UrduOCRModel(num_classes=112).to(device)
ctc_loss = nn.CTCLoss(blank=0, zero_infinity=True)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train for one epoch
def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0

    progress_bar = tqdm(dataloader, desc="📦 Training", leave=False)

    for images, labels, label_lengths in progress_bar:
        images = images.to(device)
        labels = labels.to(device)
        label_lengths = label_lengths.to(device)

        output = model(images)  # [T, B, C]
        input_lengths = torch.full(size=(output.size(1),), fill_value=output.size(0), dtype=torch.long)

        loss = criterion(output, labels, input_lengths, label_lengths)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        progress_bar.set_postfix(loss=loss.item())

    avg_loss = total_loss / len(dataloader)
    print(f"🔁 Avg training loss after 1 epoch: {avg_loss:.4f}")

# Run training loop
train_one_epoch(model, train_loader, ctc_loss, optimizer, device)
```

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from tqdm import tqdm

# Define transforms
image_transforms = transforms.Compose([
    transforms.Resize((32, 128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

class UrduOCRDataset(Dataset):
    def __init__(self, df, char_to_idx, transform=None):
        self.df = df.reset_index(drop=True)
        self.char_to_idx = char_to_idx
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # Directly use the image path without adding 'test/'
        img_path = self.df.loc[idx, 'image_path']  
        image = Image.open(img_path).convert('RGB')  # Open image directly from the path

        if self.transform:
            image = self.transform(image)

        label = self.df.loc[idx, 'label']
        label_encoded = [self.char_to_idx[c] for c in label]
        label_tensor = torch.tensor(label_encoded, dtype=torch.long)

        return image, label_tensor, len(label_encoded)  # Return label length too

# Custom collate_fn for DataLoader
def ocr_collate_fn(batch):
    images, labels, label_lengths = zip(*batch)
    images = torch.stack(images)
    labels = torch.cat(labels)
    label_lengths = torch.tensor(label_lengths, dtype=torch.long)
    return images, labels, label_lengths

# Example: Create datasets and dataloaders
train_dataset = UrduOCRDataset(train_df, char_to_idx, transform=image_transforms)
val_dataset = UrduOCRDataset(val_df, char_to_idx, transform=image_transforms)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=0, collate_fn=ocr_collate_fn)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=0, collate_fn=ocr_collate_fn)

# Visual check
batch = next(iter(train_loader))
images_batch, labels_batch, label_lengths_batch = batch
plt.imshow(images_batch[0].permute(1, 2, 0))
plt.title("Sample Image from Batch")
plt.axis("off")
plt.show()

# Optional: decode the first label
decoded_label = ''.join([idx_to_char[idx.item()] for idx in labels_batch[:label_lengths_batch[0]]])
print(f"✅ Decoded Urdu Label: {decoded_label}")

# Define the model
class UrduOCRModel(nn.Module):
    def __init__(self, num_classes):
        super(UrduOCRModel, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(128, 256, 3, padding=1), nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1), nn.ReLU(), nn.MaxPool2d((2, 1)),
            nn.Conv2d(256, 512, 3, padding=1), nn.BatchNorm2d(512), nn.ReLU(), nn.MaxPool2d((2, 1)),
            nn.Conv2d(512, 512, 2), nn.ReLU(),
        )
        self.rnn = nn.LSTM(512, 256, num_layers=2, bidirectional=True, batch_first=True)
        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        conv_out = self.cnn(x)            # [B, 512, 1, W]
        conv_out = conv_out.squeeze(2)    # [B, 512, W]
        conv_out = conv_out.permute(0, 2, 1)  # [B, W, 512]
        rnn_out, _ = self.rnn(conv_out)
        output = self.fc(rnn_out)         # [B, W, num_classes]
        return output.permute(1, 0, 2)    # [W, B, num_classes] for CTC Loss

# Instantiate model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UrduOCRModel(num_classes=112).to(device)
ctc_loss = nn.CTCLoss(blank=0, zero_infinity=True)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training and validation functions
def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    loop = tqdm(dataloader, desc="Training", leave=False)

    for images, labels, label_lengths in loop:
        images = images.to(device)
        labels = labels.to(device)
        label_lengths = label_lengths.to(device)

        output = model(images)
        input_lengths = torch.full(size=(output.size(1),), fill_value=output.size(0), dtype=torch.long)

        loss = criterion(output, labels, input_lengths, label_lengths)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    return total_loss / len(dataloader)

def validate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        loop = tqdm(dataloader, desc="Validation", leave=False)
        for images, labels, label_lengths in loop:
            images = images.to(device)
            labels = labels.to(device)
            label_lengths = label_lengths.to(device)

            output = model(images)
            input_lengths = torch.full(size=(output.size(1),), fill_value=output.size(0), dtype=torch.long)

            loss = criterion(output, labels, input_lengths, label_lengths)
            total_loss += loss.item()
            loop.set_postfix(loss=loss.item())

    return total_loss / len(dataloader)

# Train for multiple epochs
best_val_loss = float('inf')
num_epochs = 30
train_losses, val_losses = [], []

for epoch in range(num_epochs):
    print(f"\n📘 Epoch {epoch + 1}/{num_epochs}")
    train_loss = train_one_epoch(model, train_loader, ctc_loss, optimizer, device)
    val_loss = validate(model, val_loader, ctc_loss, device)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    print(f"📊 Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), "best_urdu_ocr_model.pth")
        print("💾 Saved best model!")

# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def decode_prediction(logits, idx_to_char):
    # logits shape: [W, B, C]
    pred = logits.permute(1, 0, 2)  # [B, W, C]
    pred_indices = torch.argmax(pred, dim=2)  # [B, W]

    decoded_strings = []
    for indices in pred_indices:
        prev_idx = -1
        decoded = []
        for idx in indices:
            idx = idx.item()
            if idx != prev_idx and idx != 0:  # 0 is <BLANK>
                decoded.append(idx_to_char[idx])
            prev_idx = idx
        decoded_strings.append(''.join(decoded))
    return decoded_strings
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from sklearn.model_selection import KFold
from tqdm import tqdm

# 1. Data Augmentation (in image_transforms)
image_transforms = transforms.Compose([
    transforms.RandomRotation(10),
    transforms.RandomHorizontalFlip(),
    transforms.Resize((32, 128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# 2. Modified Model with Regularization
class UrduOCRModel(nn.Module):
    def __init__(self, num_classes):
        super(UrduOCRModel, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(128, 256, 3, padding=1), nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1), nn.ReLU(), nn.MaxPool2d((2, 1)),
            nn.Conv2d(256, 512, 3, padding=1), nn.BatchNorm2d(512), nn.ReLU(), nn.MaxPool2d((2, 1)),
            nn.Conv2d(512, 512, 2), nn.ReLU(),
            nn.Dropout(0.5)  # Dropout added here
        )
        self.rnn = nn.LSTM(512, 256, num_layers=2, bidirectional=True, batch_first=True, dropout=0.3)  # Dropout in RNN
        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        conv_out = self.cnn(x)            # [B, 512, 1, W]
        conv_out = conv_out.squeeze(2)    # [B, 512, W]
        conv_out = conv_out.permute(0, 2, 1)  # [B, W, 512]
        rnn_out, _ = self.rnn(conv_out)
        output = self.fc(rnn_out)         # [B, W, num_classes]
        return output.permute(1, 0, 2)    # [W, B, num_classes] for CTC Loss

# 3. Cross-validation and Training with Early Stopping and LR Scheduler
def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0

    progress_bar = tqdm(dataloader, desc="📦 Training", leave=False)

    for images, labels, label_lengths in progress_bar:
        images = images.to(device)
        labels = labels.to(device)
        label_lengths = label_lengths.to(device)

        output = model(images)  # [T, B, C]
        input_lengths = torch.full(size=(output.size(1),), fill_value=output.size(0), dtype=torch.long)

        loss = criterion(output, labels, input_lengths, label_lengths)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        progress_bar.set_postfix(loss=loss.item())

    avg_loss = total_loss / len(dataloader)
    print(f"🔁 Avg training loss after 1 epoch: {avg_loss:.4f}")
    return avg_loss

def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0

    with torch.no_grad():
        progress_bar = tqdm(dataloader, desc="📊 Validation", leave=False)

        for images, labels, label_lengths in progress_bar:
            images = images.to(device)
            labels = labels.to(device)
            label_lengths = label_lengths.to(device)

            output = model(images)
            input_lengths = torch.full(size=(output.size(1),), fill_value=output.size(0), dtype=torch.long)

            loss = criterion(output, labels, input_lengths, label_lengths)
            total_loss += loss.item()

    avg_loss = total_loss / len(dataloader)
    print(f"🔍 Avg validation loss: {avg_loss:.4f}")
    return avg_loss

# Learning Rate Scheduler
def get_lr_scheduler(optimizer, step_size=10, gamma=0.1):
    return optim.lr_scheduler.StepLR(optimizer, step_size=step_size, gamma=gamma)

# Early Stopping Logic
class EarlyStopping:
    def __init__(self, patience=5, delta=0):
        self.patience = patience
        self.delta = delta
        self.best_loss = float('inf')
        self.counter = 0
        self.early_stop = False

    def __call__(self, val_loss):
        if val_loss < self.best_loss - self.delta:
            self.best_loss = val_loss
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True

# Main Training Loop
def train_with_cv(model, df, num_epochs=30):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    k_folds = 5
    kfold = KFold(n_splits=k_folds, shuffle=True, random_state=42)

    for fold, (train_idx, val_idx) in enumerate(kfold.split(df)):
        print(f"\nTraining fold {fold + 1}/{k_folds}...")

        # DataLoader for each fold
        train_subsam = df.iloc[train_idx]
        val_subsam = df.iloc[val_idx]

        train_dataset = UrduOCRDataset(train_subsam, char_to_idx, transform=image_transforms)
        val_dataset = UrduOCRDataset(val_subsam, char_to_idx, transform=image_transforms)

        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=ocr_collate_fn)
        val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, collate_fn=ocr_collate_fn)

        # Model, optimizer, and scheduler setup
        model = UrduOCRModel(num_classes=112).to(device)
        optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
        scheduler = get_lr_scheduler(optimizer)

        # Early stopping setup
        early_stopping = EarlyStopping(patience=3, delta=0.01)

        for epoch in range(num_epochs):
            print(f"Epoch {epoch + 1}/{num_epochs}")

            # Train one epoch
            train_one_epoch(model, train_loader, ctc_loss, optimizer, device)
            
            # Evaluate the model on the validation set
            val_loss = evaluate(model, val_loader, ctc_loss, device)

            # Check for early stopping
            early_stopping(val_loss)
            if early_stopping.early_stop:
                print("Early stopping triggered!")
                break

            # Step the learning rate scheduler
            scheduler.step()

# Start training with cross-validation and improvements
train_with_cv(model, df)
```
